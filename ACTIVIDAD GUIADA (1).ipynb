### Ejercicio 1. Representa la curva de luz de cada estrella Cefeida. Para ello puedes usar cualquier tipo de software, desde Hoja de Cálculo hasta programas tipo Gnuplot, TopCat o Period04. Debes obtener una curva parecida a la que se muestra en la Figura 4 para cada estrella. Recuerda que las magnitudes se miden de forma inversiva, es decir, a mayor brillo, menor magnitud.


```python
import os
import matplotlib.pyplot as plt
import numpy as np

#Se obtienen y grafican las curvas de luz de las estrellas ceféidas
for starfile in os.listdir('./dataHST'):
    time = []
    magnitude = []
    
    for line in open('./dataHST/'+starfile, 'r'):
        lines = [i for i in line.split()]
        time.append(float(lines[0]))
        magnitude.append(float(lines[1]))
        
    plt.figure(figsize=(10,5))
    plt.title(starfile[0:len(starfile)-4])
    plt.xlabel('Tiempo (dias)')
    plt.ylabel('Magnitud aparente')
    plt.yticks(np.arange(round(min(magnitude),1), round(max(magnitude),1), step=0.1))
    plt.gca().invert_yaxis()
    plt.plot(time, magnitude, marker = 'o', c = 'g')    
    plt.savefig(starfile[0:len(starfile)-4]+'.png')
    plt.show()
```

### 4.	Calcula el diagrama Periodo-Luminosidad para las estrellas Cefeidas de M31 a partir de las columnas Icmag y P de la tabla TopCat (o Aladin). Si es necesario, utiliza la función copiar para editor de textos/copiar para Excel en la propia tabla. Compara tus resultados con el diagrama mostrado en el artículo de Joshi et al. (2003) (ver Figura 11), el resultado obtenido debería de ser muy similar.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Se cargan los datos de las estrellas de M31
data = pd.read_csv('ejer4')

data
```

```python
#Se grafica el diagrama P-L

plt.figure(figsize=(15,10))
plt.xlabel(u'log(P) (días)', fontsize=14)
plt.ylabel('Magnitud aparente en el filtro I', fontsize=14)
plt.gca().invert_yaxis()
plt.scatter(data["log(P)"], data["Icmag"], marker = 'o', c = 'grey')    


m, b = np.polyfit(data["log(P)"], data["Icmag"], 1)

#Ajuste lineal con los datos actuales
plt.plot(data["log(P)"], m*data["log(P)"] + b + 0.5, 'y--', label='_nolegend_')
plt.plot(data["log(P)"], m*data["log(P)"] + b, 'y-')
plt.plot(data["log(P)"], m*data["log(P)"] + b - 0.5, 'y--', label='_nolegend_')

#Ajuste lineal de Joshi et al.(2003)
plt.plot(data["log(P)"],  -2.96*data["log(P)"] + 23.52, 'c--', label='_nolegend_')
plt.plot(data["log(P)"],  -2.96*data["log(P)"] + 23.02, 'c')
plt.plot(data["log(P)"],  -2.96*data["log(P)"] + 22.52, 'c--', label='_nolegend_')

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(['Ajuste actual','Ajuste Joshi et al. (2003)','Estrellas'])

plt.savefig('Diagrama P-L.png')

plt.show()

print('Pendiente del ajuste lineal actual ' +str(m))
print('Constante del ajuste lineal actual ' +str(b))
```
